```python
<!doctype html>
<html lang="en">
<script src="../static/js_an/jquery-3.3.1.slim.min.js"></script>
<script src="../static/js_an/popper.min.js"></script>
<script src="../static/js_an/bootstrap.min.js"></script>

<head>
    <meta charset="utf-8">
    <meta name="viewport" content="width=device-width, initial-scale=1, shrink-to-fit=no">
    <link rel="stylesheet" href="../static/css_an/bootstrap.min.css">
    <link href="../static/../static/vendor/fontawesome-free/css/all.min.css" rel="stylesheet" type="text/css">
    <link href="https://fonts.googleapis.com/css?family=Nunito:200,200i,300,300i,400,400i,600,600i,700,700i,800,800i,900,900i" rel="stylesheet">
    <link href="../static/css/sb-admin-2.min.css" rel="stylesheet">
    <link rel="stylesheet" href="../static/css_chat/css.css">
    <style>
        small {
            font-size: smaller;
        }
        
        p.lowercaseText {
            text-transform: lowercase;
        }
        
        .round0 {
            border-radius: 12px;
            width: 100%;
            padding: 20px 20px;
        }
    </style>
    <style>
        .resetall {
            background-color: #F5f5f5;
            border: none;
            color: #007bff;
            width: 50px;
            height: 50px;
            /* text-align: center;
        margin-bottom: 0px; */
            /* padding: 12px 16px; */
            font-size: 20px;
            cursor: pointer;
        }
        
        .record {
            background-color: #F5f5f5;
            border: none;
            color: #007bff;
            width: 50px;
            height: 50px;
            /* text-align: center;
        margin-bottom: 0px; */
            /* padding: 12px 16px; */
            font-size: 20px;
            cursor: pointer;
        }
        
        .active_ex {
            background-color: #F5f5f5;
            border: none;
            color: #007bff;
            width: 50px;
            height: 50px;
            /* text-align: center;
        margin-bottom: 0px; */
            /* padding: 12px 16px; */
            font-size: 20px;
            cursor: pointer;
        }
        .resetmesse {
            background-color: #F5f5f5;
            border: none;
            color: #007bff;
            width: 50px;
            height: 50px;
            /* text-align: center;
        margin-bottom: 0px; */
            /* padding: 12px 16px; */
            font-size: 20px;
            cursor: pointer;
        }
        /* Darker background on mouse-over
      .resetall:hover {
        background-color: RoyalBlue; */
    </style>
    <title></title>
</head>

<body>
    <div id="wrapper">

        <ul class="navbar-nav sidebar sidebar-dark accordion toggled" style="background-color: #2c3e50; min-width: 100px; margin-top:0px;margin-bottom: 0px" id="accordionSidebar">
            <!-- Sidebar - Brand -->
            <a class="sidebar-brand d-flex align-items-center justify-content-center" href="index.html">
                <div class="sidebar-brand-icon rotate-n-15">
                    <i class="fas fa-laugh-wink"></i>
                </div>
                <div class="sidebar-brand-text mx-3">Fatherlife
                    <p class="lowercaseText"><small>Life of Father</small></p>

                </div>
            </a>
            <!-- Divider -->
            <hr class="sidebar-divider my-0">
            <!-- Nav Item - Dashboard -->
            <li class="nav-item active">
                <a class="nav-link" href="index.html">
                    <i class="fas fa-home"></i>

                    <span>Home</span></a>
            </li>
            <!-- Divider -->
            <hr class="sidebar-divider">
            <!-- Heading -->
            <div class="sidebar-heading">
                Interface
            </div>
            <!-- Nav Item - Charts -->
            <li class="nav-item">
                <!-- <a class="nav-link" href="charts.html" > -->
                <a class="nav-link" href="">
                    <i class="fas fa-fw fa-chart-area"></i>
                    <span>Charts</span></a>
            </li>
            <!-- Nav Item - Tables -->
            <li class="nav-item">
                <!-- <a class="nav-link" href="tables.html" > -->
                <a class="nav-link" href="">
                    <i class="fas fa-video"></i>
                    <span>Video</span></a>
            </li>
            <!-- Divider -->
            <hr class="sidebar-divider d-none d-md-block">
            <!-- Sidebar Toggler (Sidebar) -->
            <div class="text-center d-none d-md-inline">
                <button class="rounded-circle border-0" id="sidebarToggle"></button>
            </div>
        </ul>
        <div id="content-wrapper" class="d-flex flex-column">
            <!-- Main Content -->
            <div id="content">


                <!-- Begin Page Content -->
                <div class="container-fluid">


                    <!-- Content Row -->

                    <div class="row">

                        <!-- Area Chart -->
                        <div class="col-xl-8 col-lg-7" style="padding-left: 0px;padding-right: 0px;">

                            <div class="card shadow" style="height: 100%;border-left-width: 0px;border-right-width: 0px;border-top-width: 0px;border-bottom-width: 0px;">
                                <!-- Card Header - Dropdown -->
                                <link href="../static/css_an_chat/bootstrap.min.css" rel="stylesheet">
                                <link href="../static/css_chat/bootstrap.min.css" rel="stylesheet" id="bootstrap-css">

                                <!DOCTYPE html>
                                <html class=''>

                                <head>
                                    </script>
                                    <link href='https://fonts.googleapis.com/css?family=Source+Sans+Pro:400,600,700,300' rel='stylesheet' type='text/css'>

                                    <script>
                                        try {
                                            Typekit.load({
                                                async: true
                                            });
                                        } catch (e) {}
                                    </script>
                                    <link rel='stylesheet prefetch' href='https://cdnjs.cloudflare.com/ajax/libs/font-awesome/4.6.2/css/font-awesome.min.css'>
                                    <link rel='stylesheet prefetch' href='../static/css_chat/reset.min.css'>

                                    <link rel="stylesheet" href="../static/css_chat/css.css">
                                    <div id="frame">

                                        <div id="sidepanel">

                                            <div id="profile">
                                                <div class="wrap">
                                                    <img id="profile-img" src="../static/img/image.jpg" class="online" alt="" />
                                                    <p>An Tri</p>
                                                    <div id="status-options">
                                                        <ul>
                                                            <li id="status-online" class="active"><span class="status-circle"></span>
                                                                <p>Online</p>
                                                            </li>
                                                            <li id="status-away"><span class="status-circle"></span>
                                                                <p>Away</p>
                                                            </li>
                                                            <li id="status-busy"><span class="status-circle"></span>
                                                                <p>Busy</p>
                                                            </li>
                                                            <li id="status-offline"><span class="status-circle"></span>
                                                                <p>Offline</p>
                                                            </li>
                                                        </ul>
                                                    </div>
                                                </div>
                                            </div>
                                            <div id="search" style="margin-top: 10px;margin-bottom: 20px;">
                                                <label for=""><i class="fa fa-search" aria-hidden="true"></i></label>
                                                <!-- <input id="search_contact" style="font-size:13px" type="text" placeholder="Search contacts ..." disabled> -->
                                                <input id="search_contact" style="font-size:13px" type="text" placeholder="Search contacts ...">
                                            </div>
                                            <div id="contacts">
                                                <ul>
                                                    <!-- <li id="Eng_bot" class="contact active" disabled> -->
                                                    <li id="Eng_bot" class="contact active">
                                                        <div class="wrap">
                                                            <span class="contact-status online"></span>
                                                            <img src="../static/img/robot2.jpg" alt="" />
                                                            <div class="meta">
                                                                <p class="name">Eng Bot</p>
                                                                <p class="preview">Hello, I'm FatherLife Bot</p>
                                                            </div>

                                                        </div>
                                                    </li>
                                                </ul>

                                            </div>

                                        </div>
                                        
                                        <div id="robot1" style="display:block" class="content">
                                          <h1>FATHERLIFE BOT<h1>
                                            <div class="contact-profile" style="margin-top:50px">
                                                <img src="../static/img/image.jpg" alt="" />
                                                <p>An Tri</p>

                                                <div class="social-media" style="height: 60px;">
                                            <button style="border: none" class="active_ex" id="active_ex_0"><i class="fa fa-futbol-o"></i></button>
                                                    
                                                    <button style="border: none" class="record" id="recordButton" onmousedown="mouseDown()" onmouseup="mouseUp()"><i class="fa fa-microphone" style="margin-left: 11px;margin-right:11px;
                                        "></i></button>

                                                        
                                                    <button style="border: none" class="resetmesse" id="reset_messe"><i class="fa fa-eraser" style="margin-left: 11px;margin-right:11px;
                                "></i></button>

                                                    <button style="border: none" class="resetall" id="closeall"><i class="fa fa-close" style="margin-left: 11px;margin-right:11px;
                            "></i></button>

                                                </div>
                                            </div>
                                            <div class="messages">
                                                <ul>

                                                    <li class="replies">
                                                        <img src="../static/img/robot2.jpg" alt="" />

                                                        <p style="font-size:15px;min-width: 500px">Hello, I'm FatherLife Bot, created for the purpose of fitness training.
                                                            <br> Please confirm the face and talk to me.</p>
                                                        <!-- <p>
                                              <button style="margin-right:10px;" class="resetmesse" id="reset_messe">CLEAR</button>
                          <button style="margin-right:10px;" class="resetall" id="closeall">CLOSE</button>
                          <button style="margin-right:10px;" class="resetmesse" id="reset_messe">CLEAR</button>
                          <button style="margin-right:10px;" class="resetall" id="closeall">CLOSE</button>
                          <button style="margin-right:10px;" class="resetmesse" id="reset_messe">CLEAR</button>
                          <button style="margin-right:10px;" class="resetall" id="closeall">CLOSE</button>
                                            </p> -->
                                                    </li>
                                                    <!--       
                              <li class="sent">
                                  <img src="../static/img/image.jpg" alt="" />
                                  <p>Hello!</p>
                                </li>
                                        
                              <li class="replies">
                                <img src="../static/img/robot2.jpg" alt="" />
                    
                                <p style="max-width: 1000px" class="embed-responsive embed-responsive-16by9">
                                <iframe id="video_show" class="embed-responsive-item" src="testvideo.mp4"></iframe></p>
                        
                                            </li> -->

                                                </ul>
                                            </div>

                                            <div class="message-input">
                                                <div class="wrap">
                                                    <input type="text" id="text_messenger" style="margin-left: 10px" placeholder="Write your message..." />
                                                    <!-- <input type="text" id="text_messenger" style="margin-left: 10px" placeholder="Write your message..." disabled/> -->

                                                    <!-- <button id="send_button" disabled style="margin-right:10px;" class="submit" >Send</button> -->
                                                    <button id="send_button" style="margin-right:10px;" class="submit">Send</button>
                                                </div>
                                            </div>



                                        </div>

                                    </div>
                            </div>
                        </div>

                        <!-- Pie Chart -->
                        <div class="col-xl-4 col-lg-5">
                            <div class="card shadow mb-4">
                                <!-- Card Header - Dropdown -->
                                <div style="display: block">
                                    <div class="card-header py-3 d-flex flex-row align-items-center justify-content-between">
                                        <h6 class="m-0 font-weight-bold text-primary">Video realtime</h6>

                                    </div>
                                    <!-- Card Body -->
                                    <div class="card-body">
                                        <!-- <div class="chart-pie pt-4 pb-2"> -->

                                        <p class="embed-responsive embed-responsive-16by9">
                                            <!-- <iframe  src="testvideo.mp4"></iframe> -->
                                            <video id="video" width="100%" height="100%" style="display: block" autoplay></video>
                                        </p>
                                    </div>
                                </div>

                            </div>

                            <div class="card shadow mb-4">
                                <!-- Card Header - Dropdown -->
                                <div class="card-header py-3 d-flex flex-row align-items-center justify-content-between">
                                    <h6 class="m-0 font-weight-bold text-primary">Video detect</h6>
                                </div>
                                <!-- Card Body -->
                                <div class="card-body">
                                    <!-- <div class="chart-pie pt-4 pb-2"> -->

                                    <!-- <p  class="embed-responsive embed-responsive-16by9"> -->
                                    <p style="text-align:center">
                                        <img id="photo" class='video' style="display: inline;" src="">
                                        <!-- <iframe  src="testvideo.mp4"></iframe> -->
                                    </p>
                                </div>

                            </div>
                            <div class="card shadow mb-4">
                                <!-- Card Header - Dropdown -->
                                <div class="card-header py-3 d-flex flex-row align-items-center justify-content-between">
                                    <h6 class="m-0 font-weight-bold text-primary">Resuft example</h6>

                                </div>
                                <!-- Card Body -->
                                <div class="card-body">
                                    <p id="resulf_0" style="text-align: center; font: 2em sans-serif;display: block;">
                                        <b>-------------------</b><br><b>---</b></p>
                                    <p class="round0" id="resulf_1" style="color: white;text-align: center; font: 1em sans-serif;display: none;
                        background-color: green"><b>YOUR RESULTS</b>
                                        <br><br> THE NUMBER OF ACTION <span id="result_count_2"></span>
                                        <br><br> SCORE: <span id="result_point"></span>
                                    

                                    </p>

                                </div>
                            </div>
                            <div class="card shadow mb-4">
                                <!-- Card Header - Dropdown -->
                                <div class="card-header py-3 d-flex flex-row align-items-center justify-content-between">
                                    <h6 class="m-0 font-weight-bold text-primary">Check person</h6>

                                </div>
                                <!-- Card Body -->
                                <div class="card-body">
                                    <p class="round0" id="detect_people_0" style="color: white;text-align: center; font: 2em;display: block;
                        background-color: gray"><b>NO DETECT FACE
          <br><br> SYSTEM IS LOCKED
        </b> </p>
                                    <p class="round0" id="detect_people_2" style="color: white;text-align: center; font: 2em;display: none;
                        background-color: red"><b>NO DETECT PEOPLE
          <br><br> WARNING NO PEOPLE
        </b> </p>
                                    <p class="round0" id="detect_people_1" style="color: white;text-align: center; font: 2em;display: none;
                        background-color: green"><b>DETECTED FACE</b>
                                        <br><br> NAME: <span id="check_people"></span>
                                        <br><br> EMOTION: <span id="emotion_people"></span>
                                        <br><br> SYSTEM OPEN

                                    </p>


                                    <audio style="display: none" id="audio_set" autoplay controls>
        <source id="audio_chatbot" src="../static/audio/confim.mp3" type="audio/mp3">
      </audio>

                                    <input type="text" name="check_send_camera" id="check_send_camera" style="display: none" value="0" />
                                    <!-- <span>check_send_camera</span> -->
                                    <input type="text" name="select_option" id="select_option" style="display: none" value="0" />
                                    <!-- <span>select_option</span> -->
                                    <input type="text" name="select_theduc" id="select_theduc" style="display: none" value="0" />
                                    <!-- <p>select_theduc</p> -->
                                    <input type="text" name="select_time" id="select_time" style="display: none" value="0" />
                                    <!-- <p>select_time</p> -->


                                    <input type="text" name="check_label_old" id="check_label_old" style="display: none" value="0" />
                                    <!-- <span>check_label_old</span> -->
                                    <input type="text" name="count_label" id="count_label" style="display: none" value="0" />
                                    <!-- <p>count_label</p> -->
                                    <input type="text" name="check_human" id="check_human" style="display: none" value="0" />
                                    <!-- <span>check_human</span> -->
                                    <input type="text" name="time_human" id="time_human" style="display: none" value="0" />
                                    <!-- <p>time_human</p> -->
                                    <input type="text" name="time_action" id="time_action" style="display: none" value="0" />
                                    <!-- <p>time_action</p> -->
                                    <div id="formats" style="display: none">Format: start recording to see sample rate</div>
                                    <ol id="recordingsList"></ol>
                                </div>


                            </div>
                        </div>
                    </div>

                </div>
                <!-- /.container-fluid -->

            </div>
            <!-- End of Main Content -->

        </div>
    </div>
    <!-- Bootstrap core JavaScript-->
    <script src="../static/js/app.js"></script>
    <script src="../static/vendor/jquery/jquery.min.js"></script>
    <script src="../static/vendor/bootstrap/js/bootstrap.bundle.min.js"></script>

    <!-- Core plugin JavaScript-->
    <script src="../static/vendor/jquery-easing/jquery.easing.min.js"></script>

    <!-- Custom scripts for all pages-->
    <script src="../static/js/sb-admin-2.min.js"></script>
    <!-- <script src="https://ajax.googleapis.com/ajax/libs/jquery/3.4.1/jquery.min.js"></script> -->
    </script>
    <script type="text/javascript" src="../static/js_chatbot/socket.io.min.js"></script>
    <!-- <script src="../static/js_chatbot/jquery.min.js"></script> -->
    <script src="../static/js/bot_chat.js"></script>

    <script src="https://cdn.rawgit.com/mattdiamond/Recorderjs/08e7abd9/dist/recorder.js"></script>
    <script src="https://code.jquery.com/jquery-1.11.0.js"></script>
    <script type="text/javascript" src="../static/js/socket.io.min.js"></script>
    <script>
        $(document).ready(function() {
            var socket = io.connect('wss://' + document.domain + ':' + location.port);
        })
    </script>

    <script>
        function mouseDown() {
            console.log('ghiam')
            var audio = document.getElementById('audio_set');

    var source = document.getElementById('audio_chatbot');
    console.log("recordButton change");
    source.src = "";
    source.setAttribute("src", "");
    audio.load();
                // document.getElementById("recordButton").style.color = "white";
                // document.getElementById("recordButton").style.backgroundColor = "red";
            startRecording()
        }

        function mouseUp() {
            console.log('koghiam')
                // document.getElementById("recordButton").style.color = "white";
                // document.getElementById("recordButton").style.backgroundColor = "blue";
            stopRecording()

        }
    </script>
    <script>
        //======================== SET UP SOCKETIO ======================================
        $(document).ready(function() {
                var socket = io.connect('wss://' + document.domain + ':' + location.port);
            })
            //========================  ======================================
            //======================== SEND GET CHATBOT ======================================
            // function send_chatbot(mess_send_back) {
            //   var socket = io.connect('wss://' + document.domain + ':' + location.port);
            //   socket.emit('sendmess', data = mess_send_back);
            //   socket.on('repliesmess', function (data_get) {

        //     console.log(data_get)
        //     // var messagebot = data_get;
        //     $('<li class="replies"><img src="../static/img/robot2.jpg" alt="" /><p>' + data_get + '</p></li>').appendTo($('.messages ul'));
        //     $('.contact.active .preview').html('<span>Vie Bot: </span>' + data_get);

        //     $(".messages").animate({ scrollTop: $(document).height() }, "fast");
        //   })
        // };
    </script>
    <script>
        $(".messages").animate({
            scrollTop: $(document).height()
        }, "fast");
        var socket = io.connect('wss://' + document.domain + ':' + location.port);
        console.log('socket')
        $("#profile-img").click(function() {
            $("#status-options").toggleClass("active");
        });

        $("#status-options ul li").click(function() {
            $("#profile-img").removeClass();
            $("#status-online").removeClass("active");
            $("#status-away").removeClass("active");
            $("#status-busy").removeClass("active");
            $("#status-offline").removeClass("active");
            $(this).addClass("active");

            if ($("#status-online").hasClass("active")) {
                $("#profile-img").addClass("online");
            } else if ($("#status-away").hasClass("active")) {
                $("#profile-img").addClass("away");
            } else if ($("#status-busy").hasClass("active")) {
                $("#profile-img").addClass("busy");
            } else if ($("#status-offline").hasClass("active")) {
                $("#profile-img").addClass("offline");
            } else {
                $("#profile-img").removeClass();
            };

            $("#status-options").removeClass("active");
        });

        function newMessage() {
            message = $(".message-input input").val();
            if ($.trim(message) == '') {
                return false;
            }
            $('<li class="sent"><img src="../static/img/image.jpg" alt="" /><p>' + message + '</p></li>').appendTo($('.messages ul'));
            $('.message-input input').val(null);
            $('.contact.active .preview').html('<span>You: </span>' + message);
            $(".messages").animate({
                scrollTop: $(document).height()
            }, "fast");
            // var value_select = select_option.getAttribute("value");
            // if (value_select === "5") {
            //   socket.emit('sendmess', message);
            // }
            if (message != 'active_train') {
                socket.emit('sendmess', message);
            }


            if (message === 'active_train') {
                console.log('active_train')
                check_send_camera.setAttribute("value", "4");
                select_time.setAttribute("value", 0);
                socket.emit('active_train', data = 'active_train');
                dec_0 = document.getElementById('resulf_0');
                baitheduc4();
                dec_0.style.display = 'block';
                dec_1 = document.getElementById('resulf_1');
                dec_1.style.display = 'none';
            }
            // <!-- ########################DONG CHUONG TRINH TRO VE BAN DAU########################################### -->
            // if (message === 'close_program') {
            //   console.log(message)
            //   // $(element).attr('disabled', true); 
            //   $("input[id='search_contact']").attr('disabled', true);
            //   $("button[id='send_button']").attr('disabled', true);
            //   $("button[id='reset_messe']").attr('disabled', true);
            //   $("input[id='text_messenger']").attr('disabled', true);
            //   $("li[id='Eng_bot']").attr('disabled', true);
            //   dec_0 = document.getElementById('detect_people_0');
            //   dec_0.style.display = 'block';
            //   dec_1 = document.getElementById('detect_people_1');
            //   dec_1.style.display = 'none';
            //   // var checkbot = document.getElementById('Eng_bot');
            //   // checkbot.attr('disabled');
            //   // $("input[id='search_contact']").removeAttr('disabled');

            //   // var Predict_button = document.getElementById('Eng_bot');
            //   // Predict_button.setAttribute('disabled', disabled);

            // }
            // var send_input = document.getElementById("select_option");
            // var value_select = select_option.getAttribute("value");

            // // ********************************************************
            // if ((message) === '1' && (value_select == 0)) {
            //   select1();
            //   select_option.setAttribute("value", "1");
            // }
            // // -----------------
            // if ((message) === '1' && (value_select == 1)) {
            //   baitheduc1();
            //   // socket.emit('select_td', '1');
            //   check_send_camera.setAttribute("value", "4");
            //   socket.emit('active_train', data = 'active_train1');

            // }
            // if ((message) === '2' && (value_select == 1)) {
            //   baitheduc2();
            //   // socket.emit('select_td', '2');
            //   check_send_camera.setAttribute("value", "4");
            //   socket.emit('active_train', data = 'active_train2');
            // }
            // if ((message) === '3' && (value_select == 1)) {
            //   baitheduc3();
            //   // socket.emit('select_td', '3');
            //   check_send_camera.setAttribute("value", "4");
            //   socket.emit('active_train', data = 'active_train3');
            // }
            // if ((message) === '4' && (value_select == 1)) {
            //   baitheduc4();
            //   // socket.emit('select_td', '4');
            //   check_send_camera.setAttribute("value", "4");
            //   socket.emit('active_train', data = 'active_train4');
            // }
            // if (((message) === '5' || (message) === 'cancel') && (value_select == 1)) {
            //   select5();
            //   select_option.setAttribute("value", "5");
            // }
            // // ********************************************************
            // if ((message) === '2' && (value_select == 0)) {
            //   console.log("select 2");
            //   // select_option.setAttribute("value", "2");
            //   value_select = 1;
            //   console.log(value_select);
            //   select2();
            // }

            // if ((message) === '3' && (value_select == 0)) {
            //   console.log("select 3");
            //   // select_option.setAttribute("value", "3");
            //   value_select = 1;
            //   console.log(value_select);
            //   select3();
            // }

            // if ((message) === '4' && (value_select == 0)) {
            //   console.log("select 4");
            //   // select_option.setAttribute("value", "4");
            //   value_select = 1;
            //   console.log(value_select);
            //   select4();
            // }

            // if (((message) === '5' || (message) === 'cancel') && (value_select == 0)) {
            //   console.log("select 5");
            //   select_option.setAttribute("value", "5");
            //   value_select = 1;
            //   console.log(value_select);
            //   select5();
            // }





        };
        $('.submit').click(function() {
            newMessage();
        });
        $(window).on('keydown', function(e) {
            if (e.which == 13) {
                newMessage();
                return false;
            }
        });
        socket.on('replieszzz', function(data) {
                console.log(data)
                $('<li class="replies"><img src="../static/img/robot2.jpg" alt="" /><p>' + data + '</p></li>').appendTo($('.messages ul'));
                $('.contact.active .preview').html('<span>Vie Bot: </span>' + data);
                $(".messages").animate({
                    scrollTop: $(document).height()
                }, "fast");
            })
            //# sourceURL=pen.js
    </script>


    <!-- ########################TESTCODESCRIPT########################################### -->
    <script>
        // $("#check_people").text("ANCUTE");
        // $("#emotion_people").text("vui veÌ‰");

        // emotion_people
    </script>
    <script>
        // document.getElementById('send_button').addEventListener('click', submitForm);
        // function submitForm(){
        //   console.log('$%$$@')
        // }
    </script>

    <!-- ########################TESTCODESCRIPT########################################### -->


    <!-- ########################SCRIP CAMERA########################################### -->
    <script>
        $(document).ready(function() {
            var socket = io.connect('wss://' + document.domain + ':' + location.port);
            socket.onopen = function() {
                //console.log("Openened connection to websocket");
            }
            timer = setInterval(
                function() {
                    /*****************GET IMAGE SEND***********************/
                    let imageCanvas = document.createElement('canvas');
                    let imageCtx = imageCanvas.getContext("2d");

                    imageCanvas.width = 320;
                    imageCanvas.height = 240;
                    imageCtx.drawImage(video, 0, 0, imageCanvas.width, imageCanvas.height);
                    var data_image = imageCanvas.toDataURL('image/jpeg', 1.0);
                    newblob = dataURItoBlob(data_image);
                    /*****************END GET IMAGE SEND***********************/
                    var value_check_send_camera = check_send_camera.getAttribute("value"); /* send_active_face*/
                    if (value_check_send_camera == 0) {
                        socket.emit('active_face2', data = 'active_face');
                        check_send_camera.setAttribute("value", "1");


                        $("input[id='search_contact']").attr('disabled', true);

                        $("button[id='send_button']").attr('disabled', true);

                        // $("button[id='recordButton']").attr('disabled', true);

                        $("button[id='reset_messe']").attr('disabled', true);
                        $("button[id='active_ex_0']").attr('disabled', true);
                        
                        $("input[id='text_messenger']").attr('disabled', true);
                        $("li[id='Eng_bot']").attr('disabled', true);


                        dec_0 = document.getElementById('detect_people_0');
                        dec_0.style.display = 'block';
                        dec_1 = document.getElementById('detect_people_1');
                        dec_1.style.display = 'none';
                    }

                    socket.on('done_active_face', function(data) { /* done_active_face*/
                        if (data == 'done_active') {
                            check_send_camera.setAttribute("value", "2");
                        }
                    })

                    if (value_check_send_camera == 2) /* send image face*/ {
                        socket.emit('publish', data = newblob);
                        // socket.emit('publish', data = 'newblob');

                    }

                    // socket.on('check_time', function (data) {
                    //   if (data === "1") {
                    //     select_time.setAttribute("value", 0);
                    //   }
                    // })

                    socket.on('check_time', function(data) { /* send image face*/
                        var value_select = select_time.getAttribute("value");

                        if (value_select == 0) {
                            select_time.setAttribute("value", data);
                            var value_select = select_time.getAttribute("value");
                        }
                        var time_data = parseInt(data) - parseInt(value_select)
                        if ((value_select != 0) && (time_data >= 5)) {
                            // socket.on('mqtt_message_name', function (data_3) {
                            //   // $("#check_people").text("Ancute");
                            //   $("#check_people").text(data_3.toString());
                            // })
                            socket.on('mqtt_message_name', function(data_3) {
                                    // $("#check_people").text("Ancute");
                                    $("#check_people").text(data_3.toString());
                                })
                                /* end send image face*/
                            socket.on('mqtt_message_emotion', function(data_4) {
                                $("#emotion_people").text(data_4.toString());
                            })

                            socket.on('send_audio_file', function(data_5) {

                                var audio = document.getElementById('audio_set');

                                var source = document.getElementById('audio_chatbot');
                                source.src = data_5;

                                audio.load(); //call this to just preload the audio without playing
                                audio.play();

                            })


                            select_time.setAttribute("value", 0);
                            check_send_camera.setAttribute("value", "3");


                            $("button[id='recordButton']").attr('disabled', false);
                            $("input[id='search_contact']").attr('disabled', false);
                            $("button[id='send_button']").attr('disabled', false);
                            $("button[id='reset_messe']").attr('disabled', false);
                            $("button[id='active_ex_0']").attr('disabled', false);
                            
                            $("input[id='text_messenger']").attr('disabled', false);
                            $("li[id='Eng_bot']").attr('disabled', false);
                            dec_0 = document.getElementById('detect_people_0');
                            dec_0.style.display = 'none';
                            dec_1 = document.getElementById('detect_people_1');
                            dec_1.style.display = 'block';
                            dec_0 = document.getElementById('resulf_0');
                            dec_0.style.display = 'block';
                            dec_1 = document.getElementById('resulf_1');
                            dec_1.style.display = 'none';
                            botnewMessage();
                            var audio = document.getElementById('audio_set');

                            var source = document.getElementById('audio_chatbot');
                            source.src = "../static/audio/ready.mp3";

                            audio.load(); //call this to just preload the audio without playing
                            audio.play();


                        }
                        // parseInt("10")


                    })

                    if (value_check_send_camera == 3) {
                        // socket.emit('publish', data=newblob);
                        socket.emit('deactive_face', data = 'deactive_face');
                    }

                    socket.on('done_deactive_face', function(data) {
                        if (data == 'done_deactive') {
                            check_send_camera.setAttribute("value", "10");
                        }
                    })











                    //////////////////////////////////TRAIN _____/////////////////////////////////////

                    if (value_check_send_camera == 7) {
                        socket.emit('de_active_train', data = 'deactive_train');
                        check_send_camera.setAttribute("value", "8");
                    }
                    socket.on('done_deactive_train', function(data) {
                        var theduc_select = select_theduc.getAttribute("value");
                        if ((data == 'done_deactive') && (theduc_select == 1)) {
                            check_send_camera.setAttribute("value", "9");
                            camon_baitap();
                            select_theduc.setAttribute("value", "0");



                            var count = count_label.getAttribute("value");
                            count = parseInt(parseInt(count) / 2);
                            var values = parseInt(count * 10 / 10);
                            console.log('10000000000000000000000000')
                            console.log(count)
                            console.log(values)
                            $("#result_count_2").text(count.toString());
                            $("#result_point").text(values.toString());
                            socket.emit('check_point',data=count.toString())
                            dec_0 = document.getElementById('resulf_0');
                            dec_0.style.display = 'none';
                            dec_1 = document.getElementById('resulf_1');
                            dec_1.style.display = 'block';

                            check_label_old.setAttribute("value", "0");
                            count_label.setAttribute("value", "0");

                        }
                    })

                    //     socket.on ('mqtt_message_name', function(data) {
                    //   // mqtt_message_emotion

                    //       check_send_camera.setAttribute("value", "3");
                    //       $("#check_people").text(data.toString());
                    //       socket.emit('deactive_face', data='deactive_face'); 
                    // })

                    socket.on('replies_sendmess_spe', function(data) {
                        if (data == 'replies2_sendmess_spe') {
                            var value_check_send_camera = check_send_camera.getAttribute("value");
                            if (value_check_send_camera === 10) {
                                socket.emit('active_train', data = 'active_train');
                            }
                        }
                    })

                    if (value_check_send_camera == 4) {
                        socket.emit('active_train', data = 'active_train');
                        check_send_camera.setAttribute("value", "5");
                    }


                    socket.on('done_active_train', function(data) {

                        if (data == 'done_active') {
                            socket.on('done_select_td', function(data_0) {
                                var theduc_select = select_theduc.getAttribute("value");
                                if ((data_0 == '1') && (theduc_select == 0)) {
                                    video_bai1();
                                    check_send_camera.setAttribute("value", "6");
                                    select_theduc.setAttribute("value", "1");
                                }
                                if ((data_0 == '2') && (theduc_select == 0)) {
                                    video_bai2();
                                    check_send_camera.setAttribute("value", "6");
                                    select_theduc.setAttribute("value", "1");
                                }
                                if ((data_0 == '3') && (theduc_select == 0)) {
                                    video_bai3();
                                    check_send_camera.setAttribute("value", "6");
                                    select_theduc.setAttribute("value", "1");
                                }
                                if ((data_0 == '4') && (theduc_select == 0)) {
                                    //video_bai4();
                                    check_send_camera.setAttribute("value", "6");
                                    select_theduc.setAttribute("value", "1");
                                }
                            })

                        }
                    })




                    if (value_check_send_camera == 6) {
                        // socket.emit('publish_train', data = 'newblob');
                        socket.emit('publish_train', data = newblob);
                    }


                }, 250);

            socket.on('check_time_train', function(data) {

                var value_select = select_time.getAttribute("value");

                if (value_select == 0) {
                    select_time.setAttribute("value", data);
                    var value_select = select_time.getAttribute("value");
                }
                var time_data = parseInt(data) - parseInt(value_select);
                if ((value_select != 0) && (time_data >= 30)) {
                    select_time.setAttribute("value", 0);
                    check_send_camera.setAttribute("value", "7");
                }


            })

            socket.on('check_time_canhbao', function(data) {

                var value_select = time_human.getAttribute("value");

                if (value_select == 0) {
                    time_human.setAttribute("value", data);
                    var value_select = time_human.getAttribute("value");
                }
                var time_data = parseInt(data) - parseInt(value_select);
                if ((value_select != 0) && (time_data >= 20)) {
                    time_human.setAttribute("value", 0);
                    check_send_camera.setAttribute("value", "8");

                    dec_0 = document.getElementById('detect_people_0');
                    dec_0.style.display = 'none';
                    dec_1 = document.getElementById('detect_people_1');
                    dec_1.style.display = 'none';
                    dec_2 = document.getElementById('detect_people_2');
                    dec_2.style.display = 'block';
                    
                    document.getElementById('audio_chatbot').src = "../static/audio/warning.mp3";

                    document.getElementById('audio_set').load(); //call this to just preload the audio without playing
                    document.getElementById('audio_set').play();
                }
                // parseInt("10")

            })


            socket.on('label_human', function(data) {
                var check_label = check_label_old.getAttribute("value");
                var count = count_label.getAttribute("value");
                count_label
                if ((check_label != data) && (check_label == 0)) {
                    check_label_old.setAttribute("value", data);
                }
                if ((check_label != data) && (check_label != 0)) {
                    check_label_old.setAttribute("value", data);
                    count = parseInt(count) + 1;
                    count_label.setAttribute("value", count);
                }


            })
            socket.on('mqtt_message', function(data) {
                var target = document.getElementById("photo");
                var arrayBufferView = new Uint8Array(data);
                var vc = 'data:image/png;base64,' + encode(arrayBufferView);
                // console.log(data)
                target.onload = function() {
                    window.webkitURL.revokeObjectURL(vc);
                };
                target.src = vc;
            })



            function dataURItoBlob(dataURI) {
                // convert base64/URLEncoded data component to raw binary data held in a string
                var byteString;
                if (dataURI.split(',')[0].indexOf('base64') >= 0)
                    byteString = atob(dataURI.split(',')[1]);
                else
                    byteString = unescape(dataURI.split(',')[1]);

                // separate out the mime component
                var mimeString = dataURI.split(',')[0].split(':')[1].split(';')[0];

                // write the bytes of the string to a typed array
                var ia = new Uint8Array(byteString.length);
                for (var i = 0; i < byteString.length; i++) {
                    ia[i] = byteString.charCodeAt(i);
                }

                return new Blob([ia], {
                    type: mimeString
                });
            }

            function arrayBufferToBase64(buffer) {
                var binary = '';
                var bytes = new Uint8Array(buffer);
                var len = bytes.byteLength;
                for (var i = 0; i < len; i++) {
                    binary += String.fromCharCode(bytes[i]);
                }
                return window.btoa(binary);
            }

            function encode(input) {
                var keyStr = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+/=";
                var output = "";
                var chr1, chr2, chr3, enc1, enc2, enc3, enc4;
                var i = 0;

                while (i < input.length) {
                    chr1 = input[i++];
                    chr2 = i < input.length ? input[i++] : Number.NaN; // Not sure if the index 
                    chr3 = i < input.length ? input[i++] : Number.NaN; // checks are needed here

                    enc1 = chr1 >> 2;
                    enc2 = ((chr1 & 3) << 4) | (chr2 >> 4);
                    enc3 = ((chr2 & 15) << 2) | (chr3 >> 6);
                    enc4 = chr3 & 63;

                    if (isNaN(chr2)) {
                        enc3 = enc4 = 64;
                    } else if (isNaN(chr3)) {
                        enc4 = 64;
                    }
                    output += keyStr.charAt(enc1) + keyStr.charAt(enc2) +
                        keyStr.charAt(enc3) + keyStr.charAt(enc4);
                }
                return output;
            }
        });
    </script>
    <!-- ########################SCRIP CAMERA########################################### -->
    <script>
        window.addEventListener("DOMContentLoaded", function() {
            var video = document.getElementById('video');
            var mediaConfig = {
                video: true
            };
            var errBack = function(e) {
                //console.log('An error has occurred!', e)
            };
            if (navigator.mediaDevices && navigator.mediaDevices.getUserMedia) {
                navigator.mediaDevices.getUserMedia(mediaConfig).then(function(stream) {
                    video.srcObject = stream;
                    video.play();
                });
            } else if (navigator.getUserMedia) { // Standard
                navigator.getUserMedia(mediaConfig, function(stream) {
                    video.src = stream;
                    video.play();
                }, errBack);
            } else if (navigator.webkitGetUserMedia) { // WebKit-prefixed
                navigator.webkitGetUserMedia(mediaConfig, function(stream) {
                    video.src = window.webkitURL.createObjectURL(stream);
                    video.play();
                }, errBack);
            } else if (navigator.mozGetUserMedia) { // Mozilla-prefixed
                navigator.mozGetUserMedia(mediaConfig, function(stream) {
                    video.src = window.URL.createObjectURL(stream);
                    video.play();
                }, errBack);
            }
        }, false);
    </script>

    <!-- ########################SCRIP CAMERA########################################### -->

    <!-- <script>
function change_bot_eng() {
	
		var chage_bot_eng = document.getElementById("robot2");
		chage_bot_eng.style.display="block";
		var chage_bot_vie = document.getElementById("robot1");
		chage_bot_vie.style.display="none";
		var active_eng = document.getElementById("Eng_bot");
		active_eng.class="contact active";
		var active_vie = document.getElementById("Viebot");
		active_vie.class="contact";
    };
</script> -->
    <script src="https://code.jquery.com/jquery-1.11.0.js"></script>
    <script>
        $('.active_ex').click(function() {
            console.log('active_train')
                check_send_camera.setAttribute("value", "4");
                select_time.setAttribute("value", 0);
                socket.emit('active_train', data = 'active_train');
                dec_0 = document.getElementById('resulf_0');
                baitheduc4();
                dec_0.style.display = 'block';
                dec_1 = document.getElementById('resulf_1');
                dec_1.style.display = 'none';

        })
          
        $('.resetmesse').click(function() {
            console.log('click')
            $('.messages').remove();
            $('.message-input:last').before('<div class="messages"><ul></ul></div>');

        })
          
        
        $('.resetall').click(function() {
            console.log('resetall')
            $("input[id='search_contact']").attr('disabled', true);
            $("button[id='send_button']").attr('disabled', true);
            $("button[id='reset_messe']").attr('disabled', true);
            $("button[id='active_ex_0']").attr('disabled', true);
            $("input[id='text_messenger']").attr('disabled', true);
            $("li[id='Eng_bot']").attr('disabled', true);
            dec_0 = document.getElementById('detect_people_0');
            dec_0.style.display = 'block';
            dec_1 = document.getElementById('detect_people_1');
            dec_1.style.display = 'none';
            dec_2 = document.getElementById('detect_people_2');
            dec_2.style.display = 'none';
            dec_0 = document.getElementById('resulf_0');
            dec_0.style.display = 'block';
            dec_1 = document.getElementById('resulf_1');
            dec_1.style.display = 'none';




            check_send_camera.setAttribute("value", "1");
            socket.emit('active_face2', data = 'active_face');
            $('.messages').remove();
            $('.message-input:last').before('<div class="messages"><ul><li class="replies"><img src="../static/img/robot2.jpg" alt="" /><p style="font-size:15px;min-width: 500px">Hello, Im FatherLife Bot, created for the purpose of fitness training.<br> Please confirm the face and talk to me.</p> </li></ul></div>');
        })
    </script>




    <script>
var socket = io.connect('wss://' + document.domain + ':' + location.port);
socket.on('send_process_speech', function(data) {
var audio = document.getElementById('audio_set');

//var source = document.getElementById('audio_chatbot');
//source.src = data;
document.getElementById('audio_chatbot').src = data;

document.getElementById('audio_set').load(); //call this to just preload the audio without playing
document.getElementById('audio_set').play();})
          
</script> 
</body>



</html>
```
